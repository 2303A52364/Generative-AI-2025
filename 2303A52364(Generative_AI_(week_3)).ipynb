1) Write Python code without using any libraries to find the value of x at which the
function f(x) shown in equation (1) has minimum value. Consider Gradient Descent Algorithm.
f(x) = 5x
4 + 3x
2 + 10

```python
# Gradient Descent algorithm to find the minimum of the function f(x)
# f(x) = 5x^4 + 3x^2 + 10

# Define the function f(x)
def f(x):
    return 5 * x**4 + 3 * x**2 + 10

# Define the derivative of the function f'(x)
def df(x):
    return 20 * x**3 + 6 * x

# Gradient Descent function
def gradient_descent(learning_rate, initial_guess, num_iterations):
    x = initial_guess

    # Perform gradient descent for a specified number of iterations
    for _ in range(num_iterations):
        # Compute the gradient (derivative)
        gradient = df(x)

        # Update the value of x
        x = x - learning_rate * gradient

        # Optional: print the current value of x and f(x) for each iteration
        print(f"x: {x}, f(x): {f(x)}")

    return x

# Initial guess for x
initial_guess = 0.5  # You can try different values for initial guess

# Learning rate for gradient descent
learning_rate = 0.01  # You can experiment with different learning rates

# Number of iterations to perform
num_iterations = 100

# Find the minimum using gradient descent
minimum_x = gradient_descent(learning_rate, initial_guess, num_iterations)

print(f"The value of x at which the function f(x) has the minimum value is: {minimum_x}")
```

2) Write Python code without using any libraries to find the value of x and y at which the

```python
# Gradient Descent algorithm to find the minimum of the function g(x, y)
# g(x, y) = 3x^2 + 5 * e^(-y) + 10

# Define the function g(x, y)
def g(x, y):
    import math  # We will use math.exp for exponential, but no external libraries
    return 3 * x**2 + 5 * math.exp(-y) + 10

# Define the partial derivatives (gradients) of g(x, y)
def dg_dx(x, y):
    return 6 * x  # Partial derivative with respect to x

def dg_dy(x, y):
    import math
    return -5 * math.exp(-y)  # Partial derivative with respect to y

# Gradient Descent function
def gradient_descent(learning_rate, initial_guess_x, initial_guess_y, num_iterations):
    x = initial_guess_x
    y = initial_guess_y

    # Perform gradient descent for a specified number of iterations
    for _ in range(num_iterations):
        # Compute the gradients (partial derivatives)
        gradient_x = dg_dx(x, y)
        gradient_y = dg_dy(x, y)

        # Update the values of x and y
        x = x - learning_rate * gradient_x
        y = y - learning_rate * gradient_y

        # Optional: print the current values of x, y and g(x, y) for each iteration
        print(f"x: {x}, y: {y}, g(x, y): {g(x, y)}")

    return x, y

# Initial guesses for x and y
initial_guess_x = 1.0  # You can try different values for initial guess of x
initial_guess_y = 1.0  # You can try different values for initial guess of y

# Learning rate for gradient descent
learning_rate = 0.01  # You can experiment with different learning rates

# Number of iterations to perform
num_iterations = 100

# Find the minimum using gradient descent
minimum_x, minimum_y = gradient_descent(learning_rate, initial_guess_x, initial_guess_y, num_iterations)

print(f"The values of x and y at which the function g(x, y) has the minimum value are: x = {minimum_x}, y = {minimum_y}")
```

```python
# Gradient Descent algorithm to find the minimum of the function z(x)
# z(x) = 1 / (1 + e^(-x))

# Define the function z(x)
def z(x):
    import math  # We will use math.exp for exponential
    return 1 / (1 + math.exp(-x))

# Define the derivative (gradient) of z(x)
def dz_dx(x):
    import math
    return z(x) * (1 - z(x))  # The derivative of sigmoid function

# Gradient Descent function
def gradient_descent(learning_rate, initial_guess, num_iterations):
    x = initial_guess

    # Perform gradient descent for a specified number of iterations
    for _ in range(num_iterations):
        # Compute the gradient (derivative)
        gradient = dz_dx(x)

        # Update the value of x
        x = x - learning_rate * gradient

        # Optional: print the current value of x and z(x) for each iteration
        print(f"x: {x}, z(x): {z(x)}")

    return x

# Initial guess for x
initial_guess = 0.5  # You can try different values for initial guess

# Learning rate for gradient descent
learning_rate = 0.1  # You can experiment with different learning rates

# Number of iterations to perform
num_iterations = 100

# Find the minimum using gradient descent
minimum_x = gradient_descent(learning_rate, initial_guess, num_iterations)

print(f"The value of x at which the function z(x) has the minimum value is: {minimum_x}")
```

```python
# Gradient Descent algorithm to minimize the squared error
# SE = (ExpectedOutput - PredictedOutput)^2, where PredictedOutput = M * X + C

# Define the function to calculate the predicted output
def predicted_output(M, C, X):
    return M * X + C

# Define the squared error (SE) function
def squared_error(M, C, X, expected_output):
    total_error = 0
    for i in range(len(X)):
        prediction = predicted_output(M, C, X[i])
        total_error += (expected_output[i] - prediction) ** 2
    return total_error / len(X)  # Average squared error

# Define the gradients (partial derivatives of SE with respect to M and C)
def gradients(M, C, X, expected_output):
    grad_M = 0
    grad_C = 0
    N = len(X)

    for i in range(N):
        prediction = predicted_output(M, C, X[i])
        error = expected_output[i] - prediction
        grad_M += -2 * X[i] * error
        grad_C += -2 * error

    grad_M /= N  # Average gradient with respect to M
    grad_C /= N  # Average gradient with respect to C

    return grad_M, grad_C

# Gradient Descent function to minimize squared error
def gradient_descent(X, expected_output, learning_rate, initial_M, initial_C, num_iterations):
    M = initial_M
    C = initial_C

    for _ in range(num_iterations):
        # Calculate gradients
        grad_M, grad_C = gradients(M, C, X, expected_output)

        # Update parameters M and C
        M = M - learning_rate * grad_M
        C = C - learning_rate * grad_C

        # Optional: print the current values of M, C, and the squared error
        print(f"M: {M}, C: {C}, Squared Error: {squared_error(M, C, X, expected_output)}")

    return M, C

# Example data (you can change this to any dataset)
X = [1, 2, 3, 4, 5]  # Input feature values
expected_output = [2, 4, 6, 8, 10]  # Expected output (could represent a line y = 2x)

# Initial guesses for M and C
initial_M = 0.0  # Initial guess for slope
initial_C = 0.0  # Initial guess for intercept

# Learning rate for gradient descent
learning_rate = 0.01  # You can experiment with different learning rates

# Number of iterations to perform
num_iterations = 1000  # You can adjust the number of iterations

# Find the optimal M and C using gradient descent
optimal_M, optimal_C = gradient_descent(X, expected_output, learning_rate, initial_M, initial_C, num_iterations)

print(f"The optimal values of M and C are: M = {optimal_M}, C = {optimal_C}")
```

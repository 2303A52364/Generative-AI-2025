```python
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.models import Sequential
from keras.layers import Dense,Input
from tensorflow.keras.optimizers import SGD
import pandas as pd #data manipulation

df = pd.read_csv("/content/diabetes.csv")
df
```

```python
from sklearn.preprocessing import MinMaxScaler

# Assuming 'df' is your DataFrame with numerical features
# Select numerical columns for normalization
numerical_cols = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction']  # Replace with your actual column names

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit the scaler on the numerical features and transform them
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Display the normalized DataFrame
df.head()
```

```python
x = df.drop('Outcome', axis=1)
x



```

```python
y = df['Outcome']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import Adagrad
model = Sequential()
#i/l
model.add(Input(shape=(x_train.shape[1],)))
#h/l
model.add(Dense(12, activation='swish'))
model.add(Dense(25, activation='swish'))
model.add(Dense(15, activation='swish'))
#o/l
model.add(Dense(1))

model.compile(optimizer=Adagrad(), loss='mse', metrics=['accuracy'])
history = model.fit(x_train,y_train,epochs =300,batch_size = 16,validation_split=0.2)
model.summary()
```

```python
# prompt: Calculate training and testing accuracy, build confusion matrix, also calculate recall,
# precision and F1-score. using sckit

from sklearn.metrics import confusion_matrix, classification_report

# Predict on the test set
y_pred = model.predict(x_test)

# Convert probabilities to binary predictions (assuming 0.5 as threshold)
y_pred_binary = (y_pred > 0.5).astype(int)

# Calculate training accuracy
_, train_accuracy = model.evaluate(x_train, y_train, verbose=0)
print(f"Training Accuracy: {train_accuracy * 100:.2f}%")

# Calculate testing accuracy
_, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
print(f"Testing Accuracy: {test_accuracy * 100:.2f}%")


# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_binary)
print("Confusion Matrix:\n", cm)

# Classification Report (includes precision, recall, F1-score)
print(classification_report(y_test, y_pred_binary))
```
